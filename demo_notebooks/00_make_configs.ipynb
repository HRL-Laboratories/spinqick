# Creating spinqick configuration files

Before running experiments with spinqick, you'll need a few configuration files.  Here we will demonstrate how to generate and edit them

```python
from spinqick.make_config import make_default_configs
```

The first step is to set a few environment variables which will define the desired file locations of spinqick configuration files, and the directory where you'd like to autosave data.  You can set these variables in a command line,  in Windows it would look like this:

`setx SPINQICK_DATA_DIRECTORY "your directory path"`

`setx SPINQICK_HARDWARE_CONFIG "your hardware config path"`

`setx SPINQICK_EXPERIMENT_CONFIG "your experiment config path"`

`setx SPINQICK_FILTER_CONFIG "your filter config path"`



The `make_default_configs` function will look for these env vars and generate default configs if desired. Set the arguements to be true to generate default files for hardware_config, dot_experiment_config and filter_config.

```python
make_default_configs(make_filter_cfg=True, make_experiment_cfg=True, make_hw_cfg=True, overwrite=False)
```
