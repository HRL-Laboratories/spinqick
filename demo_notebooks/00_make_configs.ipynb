# Creating spinqick configuration files

Before running experiments with spinqick, you'll need a few configuration files.  Here we will demonstrate how to generate and edit them

```python
from spinqick.make_config import make_default_filesettings, FSETTING_DIR
import os
```

The `make_default_filesettings` function will create a ~.\spinqick directory and add a default settings config file.  If the arguements make_experiment_cfg or make_hw_cfg are true, this also copies a default hardware_config and experiment_config from the spinqick repo into the directory specified by `cfg_dir`.  The user can also set the `data_dir` parameter to a directory where they'd like to save experimental data.  

```python
make_default_filesettings(cfg_dir=FSETTING_DIR, data_dir=os.path.join(FSETTING_DIR, "/data"), make_experiment_cfg=True, make_hw_cfg=True)
### run this to use default paths:
# make_default_filesettings()
```

To access or edit these settings later, you can use the classes within qick.settings

```python
from spinqick import settings
```

```python
fs = settings.FileSettings()
fs.data_directory = os.path.expanduser("~/data")
fs.save_settings() # don't forget to save settings
fs.data_directory = os.path.join(FSETTING_DIR, "/data") # changing settings back to default
fs.save_settings()

ds = settings.DacSettings()
ds.t_min_slow_dac = 1.5
ds.save_settings()
```
